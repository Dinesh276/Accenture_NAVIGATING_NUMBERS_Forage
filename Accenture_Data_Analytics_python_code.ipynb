# Combining Data set

```python
df=pd.merge(reaction, content, on='ID', how="outer",validate='m:1' )
df=pd.merge(df, reactiontype, on='reaction_type', how="outer",validate='m:1' )
```

```python
df['Datetime'] = pd.to_datetime(df['Datetime'], format='%Y-%m-%d %H:%M:%S')
```

```python
ans=pd.DataFrame(df.groupby(df['Category']).Score.sum().nlargest(5))
```

```python
ans.reset_index(level=0, inplace=True)
ans
```

```python
ans.to_csv('top5_content.csv')
```

```python
result=pd.merge(ans, df, on='Category', how="inner",validate='1:m' )
```
